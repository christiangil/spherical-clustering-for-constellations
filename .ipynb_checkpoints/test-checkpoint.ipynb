```python
def normalize(a):
    return (a - min(a)) / (max(a)-min(a))

def weighted_cosine_dists(center):
    center = int(center)
    dists = np.zeros(n_stars)
    for i in range(n_stars):
        val = np.dot(pos.iloc[center], pos.iloc[i])
        if val > 0.999:
            dists[i] = 0.
        else:
            # dists[i] = dark.iloc[center] * dark.iloc[i] * np.arccos(val)
            dists[i] = np.arccos(val)
    return dists
```
